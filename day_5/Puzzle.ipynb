--- Day 5: Supply Stacks ---
The expedition can depart as soon as the final supplies have been unloaded from the ships. Supplies are stored in stacks of marked crates, but because the needed supplies are buried under many other crates, the crates need to be rearranged.

The ship has a giant cargo crane capable of moving crates between stacks. To ensure none of the crates get crushed or fall over, the crane operator will rearrange them in a series of carefully-planned steps. After the crates are rearranged, the desired crates will be at the top of each stack.

The Elves don't want to interrupt the crane operator during this delicate procedure, but they forgot to ask her which crate will end up where, and they want to be ready to unload them as soon as possible so they can embark.

They do, however, have a drawing of the starting stacks of crates and the rearrangement procedure (your puzzle input). For example:

In this example, there are three stacks of crates. Stack 1 contains two crates: crate Z is on the bottom, and crate N is on top. Stack 2 contains three crates; from bottom to top, they are crates M, C, and D. Finally, stack 3 contains a single crate, P.

Then, the rearrangement procedure is given. In each step of the procedure, a quantity of crates is moved from one stack to a different stack. In the first step of the above rearrangement procedure, one crate is moved from stack 2 to stack 1, resulting in this configuration:

![hello](input2.png)


In the second step, three crates are moved from stack 1 to stack 3. Crates are moved one at a time, so the first crate to be moved (D) ends up below the second and third crates:

![hello](input3.png)


Then, both crates are moved from stack 2 to stack 1. Again, because crates are moved one at a time, crate C ends up below crate


![hello](input4.png)


Finally, one crate is moved from stack 1 to stack 2:


![hello](input5.png)

The Elves just need to know which crate will end up on top of each stack; in this example, the top crates are C in stack 1, M in stack 2, and Z in stack 3, so you should combine these together and give the Elves the message CMZ.


```python
import string
digits = set(string.digits)
def read_stacks(text):
    """Read the lines until the line with all numbers,
    read the input into "stacks" delimited by [ and ] """
    stack_lines = []
    for line in text.split("\n"):
        unique_chars = set(line.replace(" ", ""))
        common_chars = unique_chars & digits

        is_stack_end = len(unique_chars) == len(common_chars)
        stack_lines.append(line)
        # If all digits it is the last row
        if is_stack_end:
            break
    return stack_lines

def read_instructions(text):
    """Read the lines until the line with all numbers,
    read the input into "stacks" delimited by [ and ] """
    lines = text.split("\n")
    # e.g. move 1 from 2 to 1
    instructions = list(filter(lambda line: line.startswith("move"), lines))
    extracted_instructions = []
    for instruction in instructions:
        instruction_words = instruction.split()
        number = int(instruction_words[1])
        previous = int(instruction_words[3]) - 1
        to = int(instruction_words[5]) - 1
        extracted_instructions.append((number, previous, to))
    return extracted_instructions

def digest_stacks(stack_lines):
    """Digest stacks from lines of text
    
    Returns a list of lists where each element is an initial stack
    """
    collumn_number = 0
    line_length = len(stack_lines[0])
    stacks = []
    while collumn_number < line_length:
        current_column = [line[collumn_number] for line in stack_lines]
        is_stack = current_column[-1] in digits
        if is_stack:
            # Remove stack number
            current_column = current_column[:-1]
            # Remove spaces above elements
            current_column = list(filter(lambda char: char!=" ", current_column))
            stacks.append(current_column)
        collumn_number += 1
    return stacks

def apply_singlecrate_instructions(stacks, extracted_instructions):
    for instruction in extracted_instructions:
        number, previous, to = instruction
        transfered_crates = []
        for i in range(number):
            transfered_crate = stacks[previous].pop(0)
            stacks[to].insert(0, transfered_crate)
    return stacks
puzzle_input = open("input.txt").read()
stack_lines = read_stacks(puzzle_input)
stacks = digest_stacks(stack_lines)
extracted_instructions = read_instructions(puzzle_input)
single_crate_stacks = apply_singlecrate_instructions(stacks, extracted_instructions)
print("".join([stack[0] for stack in single_crate_stacks]))
```

Before the rearrangement process finishes, update your simulation so that the Elves know where they should stand to be ready to unload the final supplies. After the rearrangement procedure completes, what crate ends up on top of each stack?

```python
def apply_multicrate_instructions(stacks, extracted_instructions):
    for instruction in extracted_instructions:
        number, previous, to = instruction
        transfered_crates = stacks[previous][:number]
        stacks[previous] = stacks[previous][number:]
        stacks[to] = transfered_crates + stacks[to]
    return stacks

puzzle_input = open("input.txt").read()
stack_lines = read_stacks(puzzle_input)
stacks = digest_stacks(stack_lines)
extracted_instructions = read_instructions(puzzle_input)
multi_crate_stacks = apply_multicrate_instructions(stacks, extracted_instructions)
print("".join([stack[0] for stack in multi_crate_stacks]))
```
